### Machine Learning
ML o Aprendizaje Automático es un campo de la informática relacionado con la capacidad de los sistemas computacionales de aprender sin ser programados explícitamente.

El ML permite construir modelos que observan datos como un conjunto de características. Estos datos se agrupan automáticamente según patrones detectados.

Los algoritmos de ML inspirados en el proceso de aprendizaje humano, aprenden de manera iterativa y permiten descubrir información.

El ML cubre la parte estadística de la inteligencia artificial. Enseña a resolver problemas observando cientos o miles de ejemplos, aprendiendo de ellos, y luego utilizando esa experiencia para resolver el mismo problema en nuevas situaciones.

![](img/ml.jpg)

Para evaluar el modelo de forma objetiva, se divide el conjunto de datos en:

- Conjunto de entrenamiento: para que el modelo aprenda.
- Conjunto de pruebas: para medir su precisión de manera independiente.
