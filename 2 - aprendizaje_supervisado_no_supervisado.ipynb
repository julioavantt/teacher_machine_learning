### Supervisado y no supervisado
El aprendizaje  puede ser supervisado o no supervisado.

##### Supervisado
Enseñamos al modelo entrenando con algunos datos de un conjunto de datos etiquetado. Tenemos algunos datos históricos y ya conocemos la clase de cada fila. Las columnas se llaman características, que incluyen los datos. 

Los datos más utilizados son numéricos. También hay categóricos.

Hay dos tipos de técnicas de aprendizaje supervisado:

- Clasificación.
- Regresión.

##### No supervisado
No supervisamos el modelo, sino que dejamos que el modelo trabaje por su cuenta para descubrir información que puede no ser visible para el ojo humano. Se entrena con el conjunto de datos y saca conclusiones sobre datos no etiquetados. El aprendizaje no supervisado crea un entorno menos controlable ya que la máquina está creando resultados para nosotros. 

Las técnicas más utilizadas para resolverlas son:

- PCA, t-SNE, UMAP → reducción de dimensionalidad
- K-means, DBSCAN, clustering jerárquico → clustering
- Reglas de asociación (Apriori, FP-Growth) → market basket analysis
- KDE, GMM → estimación de densidad
