### Clustering
La segmentación de clientes es la práctica de dividir una base de clientes en grupos de individuos que tienen características similares. Los clientes pueden agruparse en función de varios factores, incluyendo edad, género, intereses, hábitos de gasto, etc. El requisito importante es utilizar los datos disponibles para entender e identificar cómo son similares los clientes entre sí. 

Uno de los enfoques adoptados que se puede utilizar para la segmentación de clientes es el clustering. El clustering puede agrupar datos de manera no supervisada en función de la similitud de los clientes entre sí. Particionará a tus clientes en grupos mutuamente excluyentes, por ejemplo, en tres clústeres. Los clientes en cada clúster son similares entre sí demográficamente. Ahora podemos crear un perfil para cada grupo considerando las características comunes de cada clúster. 

Clustering significa encontrar clústeres en un conjunto de datos de manera no supervisada. Un clúster es un grupo de puntos de datos u objetos en un conjunto de datos que son similares a otros objetos en el grupo, y disímiles a los puntos de datos en otros clústeres. 

Los datos no están etiquetados y el proceso no es supervisado.
