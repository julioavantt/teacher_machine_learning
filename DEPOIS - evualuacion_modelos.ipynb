### Evaluación de modelos
El objetivo es construir un modelo que prediga con precisión un caso desconocido. Para esto debemos evaluar el modelo.

Enfoques de evaluación:

##### Entrenar y probar en el mismo conjunto de datos
Seleccionamos todos los datos para entrenar y construimos el modelo y una pequeña porción del conjunto de datos, pero sin sus etiquetas para probarlo. Las etiquetas no se usan para predicción, solo como verdad fundamental. Las etiquetas se llaman valores reales del conjunto de prueba. Pasamos el conjunto de características de la porción de prueba a nuestro modelo construido y predecimos los valores objetivo. Comparamos los valores predichos por nuestro modelo con los valores reales en el conjunto de prueba. Esto indica cuán preciso es realmente nuestro modelo.

Este enfoque de evaluación probablemente tendrá una alta precisión de entrenamiento y una baja precisión fuera de muestra. La precisión de entrenamiento es el porcentaje de predicciones correctas que el modelo realiza al usar el conjunto de datos de prueba. La precisión fuera de muestra es el porcentaje de predicciones correctas que el modelo realiza en datos en los que no ha sido entrenado. Esto es fundamental porque el propósito del modelo es hacer predicciones correctas en datos desconocidos.

##### División de entrenamiento y prueba
Seleccionamos una porción de nuestro conjunto de datos para el entrenamiento. Y el resto se utiliza para la prueba. El modelo se construye en el conjunto de entrenamiento. El conjunto de características de prueba se pasa al modelo para la predicción. Los valores predichos para el conjunto de prueba se comparan con los valores reales del conjunto de prueba.

Proporciona una evaluación más precisa de la precisión fuera de muestra porque el conjunto de datos de prueba no es parte del conjunto de datos que se ha utilizado para entrenar. Es más realista para problemas del mundo real. Conocemos el resultado de cada punto de datos en el conjunto de datos pero el modelo no tiene conocimiento del resultado de estos puntos de datos. Es una prueba fuera de muestra. Sin embargo, asegúrate de entrenar tu modelo con el conjunto de prueba después, ya que no quieres perder datos potencialmente valiosos.

El problema con la división de entrenamiento y prueba es que depende en gran medida de los conjuntos de datos en los que se entrenó y probó.

El error del modelo es la diferencia entre los puntos de datos y la línea de tendencia generada por el algoritmo. 

El error absoluto medio es la media del valor absoluto de los errores. Es el error promedio.
